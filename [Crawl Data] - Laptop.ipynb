## About 1 page with many products

```python
link_list = []
for i in range(8):
    resp = re.get('https://laptopaz.vn/laptop-gaming.html?page='+str(i))
    # print(resp.status_code)
    # if resp.status_code == 200:
    soup = BeautifulSoup(resp.content)
    links = soup.find_all(class_ = 'p-name')
    print(links)
    for link in links[::2]:
      link_list.append('https://laptopaz.vn'+link['href'])
    # print(link_list)
```

```python
list_price = []
list_information = []
list_warranty = []
list_stock = []

for link in link_list:
  try:
    resp = re.get(link)
    soup = BeautifulSoup(resp.content,'html.parser')
    if (not soup.find(class_='bk-product-price')): list_price.append("None")
    else: list_price.append(soup.find(class_ = 'bk-product-price').text.strip())
    if (not soup.find(class_='bk-product-name')): list_information.append("None")
    else: list_information.append(soup.find(class_ = 'bk-product-name').text.strip())
    if (not soup.find(class_='col column-2')): list_warranty.append ("None")
    else: list_warranty.append(soup.find(class_ = 'col column-2').find('p').text[10:].strip())
    if (not soup.find(class_='detail-quantity')): list_stock.append("None")
    else: list_stock.append(soup.find(class_ = 'detail-quantity').find('span').text.strip())
  except:
    continue
```

```python
# treat the information
list_status = []
list_name = []
list_parameters = []

# get status, name, parameters
for information in list_information:
  if information[0] =='[':
    l = len(information)
    # for i in range(l):
    #   # if information[i] == '[':
    #   #   status1 = i
    #   if information[i] == ']':
    #     status2 = i
    #   if information[i] == '(':
    #     para1 = i
    #   # if information[i] == ')':
    #   #   para2 = i
    status2 = information.find(']')
    para1 = information.find('(')

    status = information[1 : status2].strip()
    name = information[status2+1 : para1].strip()
    parameters = information[para1+1 : -1]

    parameters = parameters.split(',')
    cpu = parameters[0].strip()
    ram = parameters[1].strip()
    hard = parameters[2].strip()
    gpu = parameters[3].strip()
    screen = parameters[-1].strip()

    list_status.append(status)
    list_name.append(name)
    list_parameters.append([cpu, ram, hard, gpu, screen])
  else:
    list_status.append('None')
    list_name.append('None')
    list_parameters.append(['None']*5)
```

```python
list_price1 = np.array([])
for i in range(len(df['Price'])):
  price_ = str(df['Price'][i])
  if price_ != 'None':
    if price_[-1].isdigit() == False:
      price_ = price_[:-1]
    price_ = int(price_.replace('.',''))/1000000
    list_price1 = np.append(list_price1, price_)
  else:
    list_price1 = np.append(list_price1, np.nan)
# list_price1
df['Price'] = list_price1
df = df.rename(columns={'Price': 'Price (Million VND)'})
df
```
