#### Problem 1: Column name 'Price'
* Not uniform: deal with price with tail "Đ"
* Type: Object to Int
* Change name: "Price" to "Price (Million VND)"

```python
list_price1 = np.array([])
for i in range(len(df['Price'])):
  price_ = str(df['Price'][i])
  if price_ != 'None':
    if price_[-1].isdigit() == False:
      price_ = price_[:-1]
    price_ = int(price_.replace('.',''))/1000000
    list_price1 = np.append(list_price1, price_)
  else:
    list_price1 = np.append(list_price1, np.nan)
# list_price1
df['Price'] = list_price1
df = df.rename(columns={'Price': 'Price (Million VND)'})
df
```

#### Problem 5: Column name "Status"
Set score:
* 'New 100%', 'M·ªõi 100%', 'Mới 100%': 100
* 'New Outlet': 95
* 'Like New', 'Mới 99%': 90

```python
df_raw['Status'].unique()
```

```python
df['Status'].shape
```

```python
list_ = np.array([])

for i in range(len(df['Status'])):
  ram = str(df['Status'][i])
  if ram == 'New 100%' or ram == 'M·ªõi 100%' or ram == 'Mới 100%':
    list_ = np.append(list_,100)
  elif ram == 'New Outlet':
    list_ = np.append(list_,95)
  elif ram == 'Like New' or 'Mới 99%':
    list_ = np.append(list_,90)
  else:
    list_ = np.append(list_,np.nan)

# list_
df['Status'] = list_
```
